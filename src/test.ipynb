```python
# Import necessary modules
import os

# Define the paths to the two folders
folder1 = "/home/nele_pauline_suffo/outputs/vtc"
folder2 = "/home/nele_pauline_suffo/ProcessedData/videos_superannotate_all"

# Function to get a list of filenames without extensions from a folder
def get_cleaned_filenames(folder_path, strip_after_underscore=False):
    filenames = []
    for file in os.listdir(folder_path):
        if os.path.isfile(os.path.join(folder_path, file)):
            base_name = os.path.splitext(file)[0]
            if strip_after_underscore:
                base_name = base_name.split("_")[0]
            filenames.append(base_name)
    return filenames

# Get the lists of filenames (without extensions) from both folders
list1 = get_cleaned_filenames(folder1, strip_after_underscore=True)
list2 = get_cleaned_filenames(folder2, strip_after_underscore=True)

# Convert the lists to sets for easier comparison
set1 = set(list1)
set2 = set(list2)

# Find unique elements in each list
unique_to_list1 = set1 - set2
unique_to_list2 = set2 - set1

# Print the unique elements
print("Elements only in the first list:")
print(unique_to_list1)

print("\nElements only in the second list:")
print(unique_to_list2)
```
