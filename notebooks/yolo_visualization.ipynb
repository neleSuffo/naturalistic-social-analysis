## Filtering BMN Predictions by Confidence Threshold

Great idea! Let's filter predictions by confidence and visualize only the high-confidence action segments.

```python
# Step 1: Analyze score distribution to choose a good threshold
import numpy as np
import matplotlib.pyplot as plt

# Get all predictions for the video
video_name = '365653_03'
predictions = result_dict[video_name]

# Extract all scores
all_scores = [pred['score'] for pred in predictions]
all_scores_sorted = sorted(all_scores, reverse=True)

print(f"Score Analysis for video '{video_name}':")
print("=" * 50)
print(f"Total predictions: {len(all_scores)}")
print(f"Highest score: {max(all_scores):.2e}")
print(f"Lowest score: {min(all_scores):.2e}")
print(f"Mean score: {np.mean(all_scores):.2e}")
print(f"Median score: {np.median(all_scores):.2e}")

# Show top 10 scores
print(f"\nTop 10 highest scores:")
for i, score in enumerate(all_scores_sorted[:10]):
    print(f"  {i+1}: {score:.2e}")

# Plot score distribution
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.hist(all_scores, bins=50, alpha=0.7, edgecolor='black')
plt.xlabel('Confidence Score')
plt.ylabel('Count')
plt.title('Distribution of All Scores')
plt.yscale('log')  # Log scale because scores are very small

plt.subplot(1, 2, 2)
plt.plot(range(1, len(all_scores_sorted) + 1), all_scores_sorted, 'b-', alpha=0.7)
plt.xlabel('Prediction Rank')
plt.ylabel('Confidence Score')
plt.title('Scores Ranked by Confidence')
plt.yscale('log')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Step 2: Filter predictions by confidence threshold
def filter_predictions_by_threshold(predictions, threshold):
    """Filter predictions above a confidence threshold"""
    filtered = []
    for pred in predictions:
        if pred['score'] >= threshold:
            filtered.append(pred)
    return filtered

# Choose different thresholds to experiment with
thresholds = [
    max(all_scores),  # Only the highest
    all_scores_sorted[4] if len(all_scores_sorted) > 4 else max(all_scores),  # Top 5
    all_scores_sorted[9] if len(all_scores_sorted) > 9 else max(all_scores),  # Top 10
    np.percentile(all_scores, 95),  # Top 5%
    np.percentile(all_scores, 90),  # Top 10%
]

print("Threshold Analysis:")
print("=" * 50)
for i, threshold in enumerate(thresholds):
    filtered_preds = filter_predictions_by_threshold(predictions, threshold)
    print(f"Threshold {i+1}: {threshold:.2e}")
    print(f"  Predictions kept: {len(filtered_preds)}/{len(predictions)} ({len(filtered_preds)/len(predictions)*100:.1f}%)")
    print()

# Choose a reasonable threshold (top 10 predictions)
chosen_threshold = all_scores_sorted[9] if len(all_scores_sorted) > 9 else all_scores_sorted[-1]
filtered_predictions = filter_predictions_by_threshold(predictions, chosen_threshold)

print(f" CHOSEN THRESHOLD: {chosen_threshold:.2e}")
print(f" HIGH-CONFIDENCE PREDICTIONS: {len(filtered_predictions)}")
print("\nFiltered Action Segments:")
print("=" * 30)
for i, pred in enumerate(filtered_predictions):
    segment = pred['segment']
    score = pred['score']
    duration = segment[1] - segment[0]
    print(f"  {i+1}. {segment[0]:.1f}s - {segment[1]:.1f}s ({duration:.1f}s) | Score: {score:.2e}")
```

```python
# Step 4: Analyze overlaps between predictions and ground truth
def calculate_iou(pred_segment, gt_segment):
    """Calculate Intersection over Union (IoU) between two segments"""
    pred_start, pred_end = pred_segment
    gt_start, gt_end = gt_segment
    
    # Calculate intersection
    intersection_start = max(pred_start, gt_start)
    intersection_end = min(pred_end, gt_end)
    intersection = max(0, intersection_end - intersection_start)
    
    # Calculate union
    union = (pred_end - pred_start) + (gt_end - gt_start) - intersection
    
    # Calculate IoU
    if union == 0:
        return 0
    return intersection / union

print(" DETAILED OVERLAP ANALYSIS:")
print("=" * 50)

if filtered_predictions and gt_annotations:
    print("IoU Analysis (Intersection over Union):")
    print("-" * 35)
    
    for i, pred in enumerate(filtered_predictions):
        pred_segment = pred['segment']
        pred_score = pred['score']
        
        print(f"\n Prediction {i+1}: {pred_segment[0]:.1f}s - {pred_segment[1]:.1f}s (Score: {pred_score:.2e})")
        
        best_iou = 0
        best_match = None
        
        for j, gt in enumerate(gt_annotations):
            gt_segment = gt.get('segment', [0, 0])
            gt_label = gt.get('label', 'Unknown')
            
            iou = calculate_iou(pred_segment, gt_segment)
            
            if iou > 0:
                print(f"    Overlaps with GT {j+1} ({gt_label}): IoU = {iou:.3f}")
                
                if iou > best_iou:
                    best_iou = iou
                    best_match = (j+1, gt_label)
        
        if best_match:
            print(f"    Best match: GT {best_match[0]} ({best_match[1]}) with IoU = {best_iou:.3f}")
        else:
            print(f"    No overlap with any ground truth annotation")

    # Summary statistics
    print(f"\n SUMMARY:")
    print("-" * 20)
    ious = []
    matched_predictions = 0
    
    for pred in filtered_predictions:
        pred_segment = pred['segment']
        max_iou = 0
        
        for gt in gt_annotations:
            gt_segment = gt.get('segment', [0, 0])
            iou = calculate_iou(pred_segment, gt_segment)
            max_iou = max(max_iou, iou)
        
        ious.append(max_iou)
        if max_iou > 0.1:  # Consider IoU > 0.1 as a match
            matched_predictions += 1
    
    print(f"High-confidence predictions: {len(filtered_predictions)}")
    print(f"Ground truth annotations: {len(gt_annotations)}")
    print(f"Predictions with overlap (IoU > 0.1): {matched_predictions}/{len(filtered_predictions)}")
    print(f"Average IoU: {np.mean(ious):.3f}")
    print(f"Max IoU: {max(ious):.3f}")
    
else:
    if not filtered_predictions:
        print(" No high-confidence predictions to analyze")
    if not gt_annotations:
        print(" No ground truth annotations available")
```

```python
# Step 4: Load video and create highlighted clips for high-confidence segments
import cv2
import os
from IPython.display import Video, display, HTML
import numpy as np

# Video path (adjust this to your actual video path)
video_path = f"/home/nele_pauline_suffo/ProcessedData/childlens_videos/{video_name}.MP4"

def create_highlighted_video_clips(video_path, segments, output_dir="/tmp/bmn_clips"):
    """Create video clips for high-confidence segments with visual highlights"""
    
    if not os.path.exists(video_path):
        print(f" Video file not found: {video_path}")
        return []
    
    os.makedirs(output_dir, exist_ok=True)
    
    # Open video
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)
    
    clip_paths = []
    
    for i, pred in enumerate(segments):
        segment = pred['segment']
        score = pred['score']
        start_time, end_time = segment[0], segment[1]
        
        # Calculate frame numbers
        start_frame = int(start_time * fps)
        end_frame = int(end_time * fps)
        
        # Output path for this clip
        clip_path = os.path.join(output_dir, f"clip_{i+1}_{start_time:.1f}s-{end_time:.1f}s_score_{score:.2e}.mp4")
        
        # Create video writer
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(clip_path, fourcc, fps, 
                            (int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), 
                             int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))))
        
        # Set video position to start frame
        cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
        
        frame_count = 0
        while frame_count < (end_frame - start_frame):
            ret, frame = cap.read()
            if not ret:
                break
                
            # Add highlight border (red rectangle around frame)
            cv2.rectangle(frame, (10, 10), 
                         (frame.shape[1]-10, frame.shape[0]-10), 
                         (0, 0, 255), 8)  # Red border
            
            # Add text overlay with score and time
            current_time = start_time + (frame_count / fps)
            text = f"BMN Prediction | Score: {score:.2e} | Time: {current_time:.1f}s"
            cv2.putText(frame, text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                       1, (0, 255, 255), 2, cv2.LINE_AA)  # Yellow text
            
            out.write(frame)
            frame_count += 1
        
        out.release()
        clip_paths.append(clip_path)
        print(f" Created clip {i+1}: {clip_path}")
    
    cap.release()
    return clip_paths

# Create clips for high-confidence predictions
print("Creating highlighted video clips for high-confidence predictions...")
print("=" * 60)

if len(filtered_predictions) > 0:
    clip_paths = create_highlighted_video_clips(video_path, filtered_predictions)
    
    if clip_paths:
        print(f"\n Created {len(clip_paths)} video clips!")
        print("\nTo view the clips, run the next cell...")
    else:
        print(" Could not create video clips. Check video path.")
else:
    print(" No high-confidence predictions found. Lower the threshold to see more segments.")
```
