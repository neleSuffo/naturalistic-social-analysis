### Plot confusion matrix

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Confusion matrix data (rows: True labels, columns: Predicted labels)
conf_matrix = np.array([
    [5851, 2, 63, 756],  # True: person
    [1, 1824, 3, 255],   # True: face
    [30, 1000, 6297, 343],  # True: child body parts
    [778, 228, 878, 0]   # True: background
])

# Swap axes: Predicted labels on x-axis, True labels on y-axis
conf_matrix = conf_matrix.T  # Transpose the matrix

# Compute percentages
row_sums = conf_matrix.sum(axis=1, keepdims=True)
percentages = np.round((conf_matrix / row_sums) * 100, 1)

# Labels for the confusion matrix
labels = ["person", "face", "child body parts", "background"]

# Plot confusion matrix
plt.figure(figsize=(8, 6))
ax = sns.heatmap(conf_matrix, annot=False, fmt="d", cmap="Blues", cbar=True,
                 xticklabels=labels, yticklabels=labels)

# Add text annotations
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        value = conf_matrix[i, j]
        percent = percentages[i, j]
        ax.text(j + 0.5, i + 0.5, f"{value}\n({percent}%)", ha="center", va="center", fontsize=12, color="black")

plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```
