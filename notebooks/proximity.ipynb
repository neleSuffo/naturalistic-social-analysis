```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
min_val = 7134.0
max_val = 450402.0
target_value = 27334.0  # The value to highlight

# Generate a range of values from min to max
x_values = np.linspace(min_val, max_val, 1000)

# Apply different scaling methods
linear_scaled = (x_values - min_val) / (max_val - min_val)
log_scaled = (np.log(x_values) - np.log(min_val)) / (np.log(max_val) - np.log(min_val))
sqrt_scaled = (np.sqrt(x_values) - np.sqrt(min_val)) / (np.sqrt(max_val) - np.sqrt(min_val))

# Plot the transformations
plt.figure(figsize=(8, 6))
plt.plot(x_values, linear_scaled, label="Linear Scaling", linestyle='--')
plt.plot(x_values, log_scaled, label="Logarithmic Scaling")
plt.plot(x_values, sqrt_scaled, label="Square Root Scaling")

# Highlight the target value on each curve
plt.scatter(target_value, (target_value - min_val) / (max_val - min_val), color='blue', marker='o', label="Linear (27334)")
plt.scatter(target_value, (np.log(target_value) - np.log(min_val)) / (np.log(max_val) - np.log(min_val)), color='green', marker='o', label="Log (27334)")
plt.scatter(target_value, (np.sqrt(target_value) - np.sqrt(min_val)) / (np.sqrt(max_val) - np.sqrt(min_val)), color='red', marker='o', label="Sqrt (27334)")

# Adjust x-axis limits to match min and max values
plt.xlim(min_val, max_val)
plt.xlabel("Original Value")
plt.ylabel("Scaled Value (0 to 1)")
plt.title("Comparison of Scaling Methods")
plt.legend()
plt.grid(True)
plt.show()
```
