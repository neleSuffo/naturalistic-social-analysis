```python
import pandas as pd

# Read the Excel file
df_file_names = pd.read_excel('/home/nele_pauline_suffo/ProcessedData/task_ids/quantex_at_home_data_sheet.xlsx', engine='openpyxl')

df_birthdays = pd.read_csv('/home/nele_pauline_suffo/ProcessedData/task_ids/study_subject_export.csv')
birthday_dict = df_birthdays.set_index('ID No.')['Date of Birth'].to_dict()
```

```python
# Add a new column to the DataFrame by mapping the 'id' with the dictionary
df_file_names['Birthday'] = df_file_names['ID'].map(birthday_dict)

# Select only the required columns
result = df_file_names[['file name  (generated automatically)', 'DATE', 'Birthday']]

# Display only rows where the 'Birthday' column is not NaN
results_not_nan = result.dropna(subset=['Birthday'])

results_not_nan['Birthday'] = pd.to_datetime(results_not_nan['Birthday'], format='%m/%d/%Y')

# Calculate age
results_not_nan['Age'] = (results_not_nan['DATE'] - results_not_nan['Birthday']).dt.days / 365.25

# View the result
print(results_not_nan[['file name  (generated automatically)', 'DATE', 'Birthday', 'Age']])
```

```python
age_df = results_not_nan.set_index('file name  (generated automatically)')['Age']
age_df_cleaned_all = age_df.dropna()
age_dict = age_df_cleaned_all.to_dict()
```
